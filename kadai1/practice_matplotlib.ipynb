```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
t = np.linspace(0, 10, 1000)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.abs(np.sin(t))
y4 = np.sin(t)**2

c1, c2, c3, c4 = "blue", "green", "red", "black"
l1, l2, l3, l4 = "sin", "cos", "abs(sin)", "sin**2"

ax.set_xlabel('t')
ax.set_ylabel('y')
ax.set_title(r'$\sin(x)$ and $\cos(x)$')
ax.grid()

ax.plot(t, y1, color=c1, label=l1)
ax.plot(t, y2, color=c2, label=l2)
ax.plot(t, y3, color=c3, label=l3)
ax.plot(t, y4, color=c4, label=l4)
ax.legend(loc=0)
fig.tight_layout()
plt.show()
```
